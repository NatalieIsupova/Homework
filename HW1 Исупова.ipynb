### Задача 2 (2 балла)

Посчитать 1! + 2! + 3! + 4! + ... + n! для заданного n. Решить только одним циклом

```python
def function_1(n):
    a = 1
    result = 0
    for i in range(1, n+1):
        a = a*i
        result += a
    return result

assert function_1(1) == 1
assert function_1(3) == 9

assert function_1(4) == 33 
assert function_1(5) == 153
```

### Задача 3 (2 балла)

Найти среднее арифметическое отрицательных элементов в списке

```python
def function_2(list_of_numbers):
    count = 0
    summa = 0.0
    for i in list_of_numbers:
        if i < 0:
            summa += i
            count += 1
    if count == 0:
        mean = 0
    else:
        mean = summa/count
    return mean

assert function_2([]) == 0
assert function_2([1, -1, -3, 3]) == -2

assert function_2([0, 1, 2, 3]) == 0
assert function_2([2, -2, 3, -3]) == -2.5
assert function_2([1, 0, -1, -2]) == -1.5
```

### Задача 4 (2 балла)

Найти расстояние между точками с заданными координатами в n-мерном пространстве

```python
import math
def function_3(point_1, point_2):
    return math.sqrt((point_2[0]-point_1[0])**2 + (point_2[1]-point_1[1])**2)

assert function_3((1, 0), (2, 0)) == 1

assert function_3((2, 3), (0, 2)) == 2.23606797749979
assert function_3((3, 5), (6, 4)) == 3.1622776601683795
```

### Задача 5 (3 балла)

Дан список-массив, заполненный случайным образом нулями и единицами. Найти самую длинную непрерывную последовательность единиц и определить индексы первого и последнего элементов в ней

```python
def function_4(binary_list):
    count = 0
    len_ones_sequence = 0
    begin_index = 0
    end_index = 0
    for i, j in enumerate(binary_list):
        if j == 1:
            count += 1
            if count > len_ones_sequence:
                len_ones_sequence = count
                end_index = i
                begin_index = i+1 - count
        else:
            count = 0
    return len_ones_sequence, begin_index, end_index

#Тесты
assert function_4([0, 1, 0, 1, 1, 0, 0, 0]) == (2, 3, 4)

assert function_4([0, 0, 1, 1, 1, 1, 0, 1, 1, 1, 1]) == (4, 2, 5)
assert function_4([1]) == (1, 0, 0)
assert function_4([0]) == (0, 0, 0)
```

### Задача 6 (3 балла)

Найти суммы элементов столбцов матрицы и отсортировать столбцы матрицы по возрастанию найденных сумм

```python
def function_5(matrix):
    columns = [[] for i in matrix[0]]
    for i in matrix:
        for j in range(len(i)):
            columns[j].append(i[j])
            
    sums = [sum(i) for i in columns]
    
    for i in range(len(sums)):
        for j in range(i + 1, len(sums)):
            if sums[i] > sums[j]:
                sums[i], sums[j] = sums[j], sums[i]
                columns[i], columns[j] = columns[j], columns[i]
            
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            matrix[i][j] = columns[j][i]
    return matrix

#Тесты
assert function_5([[5, 0, 1], [1, 1, 1], [2, 2, 0]]) == [[1, 0, 5], [1, 1, 1], [0, 2, 2]]

assert function_5([[2, 4, 8], [1, 2, 3], [4, 5, 0]]) == [[2, 4, 8], [1, 2, 3], [4, 5, 0]]
```

### Задача 7 (3 балла)

На вход подаётся строка S и буква w. Нужно развернуть в этой строке подстроку, заключенную между первой и последней буквой w, и вернуть получившуюся строку. Если данной буквы в строке нет, то вернуть -1. Если строка содержит только одну данную букву, то вернуть строку без изменений

```python
def function_6(S, w):
    begin = 0
    end = -1
    begin = S.find(w)
    for i in range(len(S)-1, -1, -1):
        if S[i] == w:
            end = i
            break
    if end == -1:
        return -1
    if begin == end:
        return S
    return S[0:begin+1] + S[end-1:begin:-1] + S[end:]

#Тесты
assert function_6('abcdefgefg', 'e') == 'abcdegfefg'
assert function_6('dkgdfojglsjfnc', 'j') == 'dkgdfojslgjfnc'
```

### Задача 8 (4 балла + 2 балла за оформления в виде функции, принимающей текст на русском языке)

Дан текст. Посчитать, сколько раз встречается каждое слово и вывести слова в порядке убывания количества повторений слова (при совпадении частоты встречаемости - вывести в лексикографическом порядке)

```python
import re
with open('problem_8_text.txt', 'rt', encoding='utf-8') as fl:
    dict = {}
    for i in fl:
        i = i.lower()
        for j in re.split('[., —:\n!?\ufeff]+', i):
            if j:
                if dict.get(j) is None:
                    dict[j] = 1
                else:
                    dict[j] += 1
    l = [(-i,j) for j,i in dict.items()]
    l.sort()
    l = [(-i,j) for i,j in l]
    for i,j in l:
        print(j,i)
```
